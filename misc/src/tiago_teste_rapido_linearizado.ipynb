```python
with open("temp.txt", "r") as f:
    corpus = f.read().split("==========")

print( 'There are ' + str(len(corpus)) + " contexts" + '\n')
```

```python
df = pd.DataFrame({'text':corpus})
df["id"] = ["text_" + str(i) for i in df.index]
```

```python
classes = cp.classificar(df)
```

```python
n_target = 0
for key in classes: 
    if classes[key] == "1": 
        n_target += 1

total = 0
for key in classes: 
    if classes[key] != "unlabeled": 
        total += + 1

prop_before_enrichment = n_target/total


print(f"{n_target} compras em {total} classificaçoes ")
print(f"proporção de compras: {prop_before_enrichment}")
```

```python
df["label"] = df["id"].map(classes)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.semi_supervised import LabelSpreading
import nltk

vectorizer = TfidfVectorizer(stop_words=nltk.corpus.stopwords.words('portuguese'))
X = vectorizer.fit_transform(corpus)


foo = df["label"]
foo[foo == "unlabeled"] = -1


label_spread = LabelSpreading(kernel='knn', alpha=0.8)
label_spread.fit(X.toarray(), foo.astype(int))
```

```python
output_labels = label_spread.transduction_
output_labels

df["inferred_labels"] = output_labels
```

```python
fig, axs = plt.subplots(ncols=2)

g = sns.scatterplot(x="d1", y="d2",
          hue="label",
          data=df,
          ax=axs[0]);


g2 = sns.scatterplot(x="d1", y="d2",
          hue=output_labels.astype(str),
          data=df,
          ax=axs[1]);


```
